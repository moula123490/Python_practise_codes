```python
def only_numeric_nested_list(a):
    n=[]
    for i in a:
        if type(i)==list:
            for j in i:
                if type(j)==int or type(j)==float:
                    n.append(j)
        else:
            if type(i)==int or type(i)==float:
                n.append(i)
    return n

only_numeric_nested_list([[1,2,3],["moula","babar"],45,67])
```

```python
#Simple Use-Case Example

'''Suppose you’re writing a function to register a user, but you don’t know exactly which extra information 
(like age, city, interests, etc.) different parts of your program might want to pass. You can use **kwargs:'''
def register_user(username, **kwargs):
    # Mandatory parameter
    user = {"username": username}
    
    # Add any extra keyword args to the user dict
    for key, value in kwargs.items():
        user[key] = value
    
    # Simulate saving user to database
    print("Registering user with data:", user)
    return user

# Example calls:
user1 = register_user("alice", age=25, city="Bengaluru")
user2 = register_user("bob", city="Mumbai", interests=["music", "coding"])
```

```python
# 5)  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms?

def fib_series(x):
    a,b=0,1
    for i in range(x):
        yield a
        a,b=b,a+b
gen=fib_series(9)
for i in gen:
    print(i,end = " ")


```

```python
# extract even numbers from a list using init method
# the biggest advantage from init method is we can initialize/pass different arguments for different object and here flexibility is there
# but without init method we have to pass 

class listops():
    def __init__(self,l):
        self.l=l
        return
    def listops_even(self):
        
        l1=[]
        for i in self.l:
            if i%2==0:
                l1.append(i)
        return l1
    def listops_odd(self):
        l1=[]
        l=self.l
        for i in self.l:
            if i%2!=0:
                l1.append(i)
        return l1
obj=listops([1,2,3,4,5,6])

print(obj.listops_even())
print(obj.listops_odd())
```

```python
# multilevel inheritance

#grandfather-->father-->son

class grand_pa():
    def grandpa_prop(self):
        print("i am your grandfather, you are the one who is eligible to my properties")
class father(grand_pa):
    def father_prop(self):
        print("i am your father, you are the one who is eligible to my properties")
class son(father):
    def son_prop(self):
        print("i am eligible to take all properties")

obj_son=son()
obj_son.son_prop()
obj_son.father_prop()
obj_son.grandpa_prop()
obj_father=father()
obj_father.grandpa_prop()
```

```python
#polymorphism
class teacher():
    def lecture_info(self):
        return "this lecture info is w.r.t teacher info"
class student():
    def lecture_info(self):
        return "this lecture info is w.r.t student info"
obj1=teacher()
obj2=student()

class_obj=[obj1,obj2]

def parcer(class_obj):
    for i in class_obj:
        print(i.lecture_info())
  
parcer(class_obj)
```

```python
class encapsule():
    def __init__(self,name,age):
        self.__name=name
        self.age=age
    def details(self):
        name=self.__name
        age=self.age
        return f"my name is {name} and age is {age}"
obj=encapsule('shaik',26)
print(obj._encapsule__name)# accessing the private varaiable
print(obj.details())
        
```

```python
# decorator function

def my_decorator(funct):
    def wrapper():
        print("before computation")
        funct()
        print("after computation")
    return wrapper
@my_decorator
def say_hello():
    print('hi i am shaik')
say_hello()
```

```python
class my_decorator():
    def __init__(self,func):
        self.func=func
        print("inside the init method")
    def __call__(self):
        print("something is before function")
        self.func()
        print("something is after function")
@my_decorator
def say_hello():
    print("hi shaik")
say_hello()

        
```

```python
#static method

#this is normal method
class math():
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __call__(self):
        print("hi moula")
    def add(self):
        x=self.x
        y=self.y
        return x+y
obj=math(3,5)
obj.add()  
```

```python
# this is static method

class math():
    @staticmethod
    def add(x,y):
        return x+y
obj=math()
obj.add(4,5)
```

```python
#property decorator used to access method as attribute 

class area_of_circle():
    def __init__(self,radius):
        self.radius=radius
    @property
    def area(self):
        radius=self.radius
        area=3.14*radius**2
        return area
obj=area_of_circle(5)
print("the area of the circle is ---> ",obj.area,"sq.cm")
```

```python
class student():
    def __init__(self,name,age,marks):
        self.name=name
        self.age=age
        self.marks=marks
    def task(self,task_name):
        return "task is ",task_name
    def display(self):
        print(f"name is {self.name}")
        print(f"age is {self.age}")
        
class master(student):
    def __init__(self,name,age,subject):
        marks=0
        super().__init__(name,age,marks)
        self.subject=subject
    def task(self):
        return super().task('teaching')
    def display(self):
        super().display()

#obj_stud=student("sahibzada",26,76)
#obj_stud.task("reading")

obj_master=master('ali',27,"python")
obj_master.task()
```

```python
#abstarction using abstarct methods and classes

from abc import ABC,abstractmethod
class A(ABC):
    def __init__(self,radius):
        self.radius=radius
    @abstractmethod
    def method_one(self):
        pass
    @abstractmethod
    def method_two(self):
        pass
    def method_three(self):
        print("method three")

class b(A):
    def method_one(self):
        area=3.14*self.radius**2
        print("the radius of the given number is --->",area)
    def method_two(self):
        return "inside method two"

obj=b(5)
obj.method_one()
```
